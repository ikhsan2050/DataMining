```python
y = df['Hasil IMT']
# X = df.drop(['Hasil IMT'], axis = 1)
X = df[['Jenis Kelamin', 'Umur Tahun', 'Lama Sakit', 'Tinggi', 'Berat Badan', 'IMT']]
```

```python
X['Jenis Kelamin'] = X['Jenis Kelamin'].str.replace('L', '0')
X['Jenis Kelamin'] = X['Jenis Kelamin'].str.replace('P', '1')
X['Umur Tahun'] = X['Umur Tahun'].str.replace(' Tahun', '')
X['Lama Sakit'] = X['Lama Sakit'].str.replace(' Hari', '')
X['Tinggi'] = X['Tinggi'].str.replace(' cm', '')
X['Berat Badan'] = X['Berat Badan'].str.replace(' kg', '')
# df['Diastole'] = df['Diastole'].str.replace(' Hg', '')
```

```python
import numpy as np

outliers = []
def detect_outlier(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for y in data:
        z_score = (y-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(y)
    return outliers

X['Tinggi'] = X['Tinggi'].astype(float)
outlier_tinggi = detect_outlier(X['Tinggi'])
print('Outlier Tinggi Badan = ', outlier_tinggi)
outliers.clear()

X['Berat Badan'] = X['Berat Badan'].astype(float)
outlier_berat = detect_outlier(X['Berat Badan'])
print('Outlier Berat Badan = ', outlier_berat)
outliers.clear()
```

```python
X.drop(index=outlier_tinggi, inplace=True)
X.drop(index=outlier_berat, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=100, test_size=0.3)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

model=SVC()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)

print("HASIL PREDIKSI SVM")
print(y_pred)
print("HASIL CONFUSION MATRIX")
print(confusion_matrix(y_test, y_pred))
print("HASIL AKURASI PEMODELAN SVM:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

model = GaussianNB()

# Train the classifier:
model.fit(X_train, y_train)
# Make predictions with the classifie
y_pred=model.predict(X_test)
print(y_pred)

# Evaluate label (subsets) accuracy:
print("HASIL CONFUSION MATRIX")
print(confusion_matrix(y_test, y_pred))
print("HASIL AKURASI PEMODELAN NB:", accuracy_score(y_test, y_pred))
```
